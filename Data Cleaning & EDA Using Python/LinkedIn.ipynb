## Top 10 Skills by Job Count

```python
skill_counts = []
for _, row in ski.iterrows():
    skill_abr = row["skill_abr"]
    skill_name = row["skill_name"]
    job_count = jos[jos["skill_abr"] == skill_abr]["job_id"].count()
    skill_counts.append((skill_name, job_count))
skill_counts_df = pd.DataFrame(skill_counts, columns=["Skill Name", "Job Count"])
top_skills = skill_counts_df.groupby("Skill Name")["Job Count"].sum().nlargest(10).sort_values(ascending=True)
plt.figure(figsize=(12, 8))
colors = plt.cm.Blues(np.linspace(0.4, 1.0, len(top_skills)))
bars = plt.barh(top_skills.index, top_skills.values / 1000, color=colors)  # تحويل القيم إلى آلاف
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{bar.get_width():.1f}K", va="center")
plt.title("Top 10 Skills by Job Count", fontsize=16)
plt.xlabel("Number of Jobs", fontsize=14)
plt.tight_layout()
plt.show()
```
